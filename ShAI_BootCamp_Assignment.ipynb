<h3>Descriptive Statistics begin</h3>

```python
# Select relevant columns
salary_columns = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']

# Display basic statistics
basic_stats = df[salary_columns].describe()
basic_stats
```

```python
# Calculate mode separately
mode_values = df[salary_columns].mode()
mode_values
```

```python
# Calculate range
range_values = df[salary_columns].max() - df[salary_columns].min()
range_values
```

```python
# Calculate standard deviation
std_deviation = df[salary_columns].std()
std_deviation
```

<h3>Descriptive Statistics end</h3>

<h3>Data Cleaninig begin</h3>

```python
#because they do not contain any useful information
data = df.drop(columns=['Notes','Status'])
data
```

```python
# i used fill function with mean in BasePay column based on same job in the same year 
data.loc[:, 'BasePay'] = data.groupby(['JobTitle', 'Year'])['BasePay'].transform(lambda group: group.fillna(group.mean()))
```

```python
# the missing values are limited to the year 2011 so i can't use fill with mean or median function in these blank cells ,aslo i can't drop them because these rows have Important information 
data.loc[data['Year'] == 2011, 'Benefits'] = 0
data
```

```python
#not contain any useful information
data = data[data["EmployeeName"] != "Not provided"]
data
```

```python
#these rows have unique job Title so i can't fill it with mean so will drop it becuase the Base Pay is nan  
nan_rows = data[data['BasePay'].isna()]
nan_rows
```

```python
data.dropna()
```

<h3>Data Cleaninig End</h3>


<h3>Basic Data Visualization Begin</h3>


```python
plt.figure(figsize=(10, 6))
sns.histplot(data['TotalPay'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries')
plt.xlabel('Total Pay')
plt.show()
```

```python
department_counts = data['JobTitle'].value_counts().head(10)

plt.pie(department_counts, labels=department_counts.index.tolist(), autopct='%1.1f%%', startangle=60)
plt.title('Proportion of Employees in Different Departments')
plt.show()
```

<h3>Basic Data Visualization End</h3>


<h3>Grouped Analysis Begin</h3>


```python
grouped_data = data.groupby(['Year', 'JobTitle'])
average_salary_by_group = grouped_data['TotalPay'].mean()
average_salary_by_group
```

<h3>Grouped Analysis End</h3>


<h3>Simple Correlation Analysis Begin</h3>


```python
numerical_column = 'OtherPay'  

# Calculate correlation between 'TotalPay' and OtherPay column
correlation = data['TotalPay'].corr(data[numerical_column])

# Plotting a scatter plot
sns.scatterplot(x='TotalPay', y=numerical_column, data=data)
plt.title(f'Scatter Plot: TotalPay vs. {numerical_column} (Correlation: {correlation:.2f})')
plt.xlabel('Total Pay')
plt.ylabel(numerical_column)
plt.show()
```

```python
numerical_column = 'OvertimePay'  

# Calculate correlation between 'TotalPay' and OtherPay column
correlation = data['TotalPay'].corr(data[numerical_column])

# Plotting a scatter plot
sns.scatterplot(x='TotalPay', y=numerical_column, data=data)
plt.title(f'Scatter Plot: TotalPay vs. {numerical_column} (Correlation: {correlation:.2f})')
plt.xlabel('Total Pay')
plt.ylabel(numerical_column)
plt.show()
```

```python
numerical_column = 'Benefits'  

# Calculate correlation between 'TotalPay' and OtherPay column
correlation = data['TotalPay'].corr(data[numerical_column])

# Plotting a scatter plot
sns.scatterplot(x='TotalPay', y=numerical_column, data=data)
plt.title(f'Scatter Plot: TotalPay vs. {numerical_column} (Correlation: {correlation:.2f})')
plt.xlabel('Total Pay')
plt.ylabel(numerical_column)
plt.show()
```
